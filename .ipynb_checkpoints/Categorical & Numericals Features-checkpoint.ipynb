 Reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest

```python
pd.DataFrame(categoricals['GENDER'].value_counts())
```

```python
print(categoricals['GENDER'].value_counts())
categoricals['GENDER'] = categoricals['GENDER'].fillna('F')
```

```python
df = pd.DataFrame(categoricals['GENDER'].value_counts()).reset_index()

df.columns = ['GENDER', 'count']
other_states = list(df[df['count']<2500]['GENDER'])
#other_states
def clean_state(x):
    if x in other_states:
        return 'other'
    else:
        return x
    
categoricals['GENDER'] = list(map(clean_state, categoricals['GENDER']))
```
